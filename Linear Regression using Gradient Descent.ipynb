```python
x = dataset["YearsExperience"]
```

```python
y = dataset["Salary"]
```

### Split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

* Choose random m and c value 

```python
m = random.randint(10,1000)
```

```python
c = random.randint(10,1000)
```

```python
m,c
```

* Choose specific  learning rate and no of iterations

```python
lr = 0.01
iterations = 1000
```

```python
n = float(len(x_train))
```

```python
n
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for i in range(iterations):
    
    #calculate y_pred
    y_pred_train = (m * x_train) + c
    
    
    #Caculate Derivatives
    dm =  - (2 / n) * sum(x_train * (y_train - y_pred_train))
    dc =  - (2 / n) * sum((y_train - y_pred_train))
    
    #Update m,c values
    m = m - (lr * dm)
    c = c - (lr * dc)
    
    #Caculate MSE Value
    mse = mean_squared_error(y_train,y_pred_train)
    rmse = np.sqrt(mse)
    print("Iter::",i,"MSE::",mse,"RMSE::",rmse)
    
    
    
print("MSE",mse)
print("M&C",m,c)
```
